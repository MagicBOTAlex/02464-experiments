```python
import pandas as pd

def compute_primacy_score(row):
    gen_words = str(row['generated_words']).split()
    user_words = str(row['user_input']).split()
    first_three = gen_words[:3]

    # Handle test == 8 (case-insensitive letters)
    if row['test'] == 8:
        gen_words = [w.lower() for w in gen_words]
        user_words = [w.lower() for w in user_words]
        first_three = gen_words[:3]

    if row['quiz_type'] == 'unordered':
        return sum(word in user_words for word in first_three)
    elif row['quiz_type'] == 'ordered':
        return sum(i < len(user_words) and user_words[i] == word
                   for i, word in enumerate(first_three))
    else:
        return None


def compute_recency_score(row):
    gen_words = str(row['generated_words']).split()
    user_words = str(row['user_input']).split()
    last_three = gen_words[-3:]

    if row['test'] == 8:
        gen_words = [w.lower() for w in gen_words]
        user_words = [w.lower() for w in user_words]
        last_three = gen_words[-3:]

    if row['quiz_type'] == 'unordered':
        return sum(word in user_words for word in last_three)
    elif row['quiz_type'] == 'ordered':
        start_idx = max(0, len(gen_words) - 3)
        return sum(
            i < len(user_words) and user_words[i] == word
            for i, word in enumerate(last_three, start=start_idx)
        )
    else:
        return None


def compute_nonprimacy_score(row):
    gen_words = str(row['generated_words']).split()
    user_words = str(row['user_input']).split()
    remaining = gen_words[3:]

    if row['test'] == 8:
        gen_words = [w.lower() for w in gen_words]
        user_words = [w.lower() for w in user_words]
        remaining = gen_words[3:]

    if row['quiz_type'] == 'unordered':
        return sum(word in user_words for word in remaining)
    elif row['quiz_type'] == 'ordered':
        return sum(
            i < len(user_words) and user_words[i] == word
            for i, word in enumerate(remaining, start=3)
        )
    else:
        return None


def compute_nonrecency_score(row):
    gen_words = str(row['generated_words']).split()
    user_words = str(row['user_input']).split()
    remaining = gen_words[:-3]

    if row['test'] == 8:
        gen_words = [w.lower() for w in gen_words]
        user_words = [w.lower() for w in user_words]
        remaining = gen_words[:-3]

    if row['quiz_type'] == 'unordered':
        return sum(word in user_words for word in remaining)
    elif row['quiz_type'] == 'ordered':
        return sum(
            i < len(user_words) and user_words[i] == word
            for i, word in enumerate(remaining)
        )
    else:
        return None

# Apply all four scores
df_All['primacy_score'] = df_All.apply(compute_primacy_score, axis=1)
df_All['recency_score'] = df_All.apply(compute_recency_score, axis=1)
df_All['nonprimacy_score'] = df_All.apply(compute_nonprimacy_score, axis=1)
df_All['nonrecency_score'] = df_All.apply(compute_nonrecency_score, axis=1)

# Normalize to proportions
df_All['primacy_score'] = (df_All['primacy_score'] / 3)
df_All['recency_score'] = (df_All['recency_score'] / 3)
df_All['nonprimacy_score'] = (df_All['nonprimacy_score'] / 7)
df_All['nonrecency_score'] = (df_All['nonrecency_score'] / 7)

df_All.shape
```

# Hypothesis Testing

We test the general data of everyone using a mixed-effects model. This is because the several sample by each participant makes the samples corelated and not independant. The mixed effects model is able to use all samples instead of having to aggregate the samples to have only four independent datapoints. 

The mixed-effects model assumes: 
* linearity between the predictor (score or test type) and the outcome (score percentage) 
* Equal residual variance 
* Normality of residuals
* Normality of random effects
* Observations (trials) are independent conditional on the random effect (participant)

### Testing primacy and recency effects

For this test we are testing if participants recall the first three words better than the other words (primacy effect) and if they recall the last three words better than the rest (recency effect)

the primacy, nonprimacy, recency and nonrecency scores are recorded in proportions. (not already multiplied by 100)

First we test the primacy effect

```python
import statsmodels.formula.api as smf
import pandas as pd

# Reshape to long format
df1_long = pd.melt(
    df_All_test1,
    id_vars=['owner', 'test'],
    value_vars=['primacy_score', 'nonprimacy_score'],
    var_name='score_type',
    value_name='score_value'
)

# Fit mixed model
model = smf.mixedlm(
    "score_value ~ score_type",
    data=df1_long,
    groups=df1_long["owner"]
)
result1 = model.fit()
print(result1.summary())
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
from scipy import stats

fitted = result1.fittedvalues
residuals = result1.resid

# Residuals vs fitted
plt.figure()
plt.scatter(fitted, residuals)
plt.axhline(0, color='red', ls='--')
plt.xlabel('Fitted')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted')
plt.show()

# Q-Q plot of residuals
sm.qqplot(residuals, line='45', fit=True)
plt.title('Q-Q Plot of Residuals')
plt.show()

# Histogram
plt.hist(residuals, bins=15, edgecolor='k')
plt.title('Residuals histogram')
plt.show()

# Random effects (intercepts)
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Extract random effects as a DataFrame for easy inspection
re_dict = result1.random_effects
re_df = pd.DataFrame(re_dict).T  # transpose so participants are rows
print("Random effects table:")
print(re_df)

# If only random intercepts (one column)
if re_df.shape[1] == 1:
    re_vals = re_df.iloc[:, 0]
    sm.qqplot(re_vals, line='45', fit=True)
    plt.title('Q-Q Plot for Random Intercepts')
    plt.show()

# If multiple random effects (e.g., intercept + slope)
else:
    for col in re_df.columns:
        sm.qqplot(re_df[col], line='45', fit=True)
        plt.title(f'Q-Q Plot for Random Effect: {col}')
        plt.show()

```

As the residual plot is randomly scattered around 0 without clear patterns across, the Q-Q plots roughly follow the diagonal lines, the random effects table showing low values and the histgram being bell shaped it is fairly safe to assume the assumptions hold for the mixed-effects model.

As such we can intepret the results. 
The intercept being 0.451 is the average nonprimacy score of 45.1%.
The score_type(T.primacy_score) of 0.169 is the estimated difference between the primacy and nonprimacy scores, in this case it is estimated that the three first words are recalled about 17% better than the other words. 

Because the p-value for the difference estimate is below a significance level of 0.05 the difference is statistically significant and the primacy effect is shown.

### Testing the effect of increased presentation rate 

For this test we are comparing the overall score percentage mean for test 1 which presented the words at a rate of 1 word per 2 secs. with the overall score percentage mean for test 2 which presented the words at a rate of 1 word per 0.75 secs.

We again use th mixed-effects model to compare the overall score between the two different tests.

the overall score is recorded in percent already for both tests.

```python
import statsmodels.formula.api as smf
import pandas as pd

# Combine test data into a single df with columns: participant_id (owner), test_type ('slow'/'fast'), score_percentage, ...
df_rate = pd.concat([df_All_test1.assign(test_type='slow'), df_All_test2.assign(test_type='fast')])

model = smf.mixedlm("score_percentage ~ test_type", data=df_rate, groups=df_rate["owner"])
result3 = model.fit(reml=False)   # try ML or REML; for fixed-effect tests use ML (reml=False)
print(result3.summary())
```

### Test if people only are able to 7 items in a specific sequence 

From the results know from previous expreiments (reference til experimentet beskrevet i opgave formuleringen i vores lecture notes) they found people on average remenber 7 items when having to recall the specific sequence. 
To test this we will try to calculate a confidence interval for our average number of words remembered and check wether 7 is within this interval.
if 7 is within the interval we can neither deny or confirm wheter or not seven is the true average of the population (everyone in the world).

Because we only have 4 participants and clustered samples thus not every sample being independent we again use the mixed-effects model to estimate the CI for the average score, however because it is unstable we also run a non parametric cluster bootstrap that resamples from each participant with replacement in order to compare and evaluate better.

```python
import statsmodels.formula.api as smf

# df_All_test7 should contain columns: 'score_percentage' and 'owner'
model = smf.mixedlm("score_percentage ~ 1", data=df_All_test7, groups=df_All_test7["owner"])
res = model.fit(reml=True)   # or reml=False; choice is minor for fixed-effect CI
print(res.summary())

# parameter CI (intercept is the marginal mean of the reference)
print("Intercept CI:\n", res.conf_int().loc['Intercept'])

# If you had predictors and wanted predicted marginal means, use .predict()
```

```python
import numpy as np
import pandas as pd

def cluster_bootstrap_mean(df, group_col='owner', value_col='score_percentage', n_boot=5000, random_state=0):
    rng = np.random.default_rng(random_state)
    groups = df[group_col].unique()
    B = n_boot
    boot_means = np.empty(B)
    for b in range(B):
        # sample groups with replacement
        sampled_groups = rng.choice(groups, size=len(groups), replace=True)
        # concatenate all rows for sampled groups
        sampled = pd.concat([df[df[group_col] == g] for g in sampled_groups], ignore_index=True)
        boot_means[b] = sampled[value_col].mean()
    # percentile CI
    lower, upper = np.percentile(boot_means, [2.5, 97.5])
    return boot_means.mean(), (lower, upper), boot_means

mean_boot, (ci_low, ci_high), boot_dist = cluster_bootstrap_mean(df_All_test7, n_boot=5000)
print("Cluster-bootstrap mean:", mean_boot)
print("95% percentile CI (cluster bootstrap):", (ci_low, ci_high))
```

After rounding the percent scores to word counts the mixed-effects model gives an average score of 4 words with the CI being [1;7] and the bootsrap also gives an average of 4 words but with a CI being [2;7].
As seven is included in both confidence intervals our sample does not deny that the average recall ability of a words in a sequence is 7 words but points to the average being 4 words.
